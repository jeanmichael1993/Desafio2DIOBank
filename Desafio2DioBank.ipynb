```python
def menu():
  menu = """\n
  [d] Depositar
  [s] Sacar
  [e] Extrato
  [nc] Nova Conta
  [lc] Listar Contas
  [nu] Novo Usuário
  [q] Sair """
  return menu


def deposito(valor: float, saldo: float, extrato:str):
  if valor <= 0:
    return "Valor inválido"
  extrato.append("+"+str(valor))
  saldo += valor
  print("Deposito realizado com sucesso!")
  return extrato, saldo


def saque(valor: float, saldo:float, saques_realizados: float, valor_limite_saque:float, saques_limite:int, extrato: str):
  if valor <= 0:
    return "Valor inválido"
  if valor > valor_limite_saque:
    return "Valor acima do limite"
  if saques_realizados < saques_limite:
    if saldo >= valor:
      saldo -= valor
      extrato.append("-"+str(valor))
      saques_realizados += 1
      print("Saque realizado com sucesso")
      return extrato, saldo
    else:
      return "Saldo insuficiente"
  else:
    return "Limite de saques atingido"

def extrato_func(saldo:float, extrato:str):
  print("Extrato:")
  for i in extrato:
    print(i)
  return print(f"Seu saldo é de R${saldo :.2f}")

def nova_conta(agencia: str, numero_conta: int, usuarios: str):
  cpf = input("Informe o CPF (somente número): ")
  usuario = filtrar_usuario(cpf, usuarios)
  if usuario:
    print("\nConta criada com sucesso!")
    return {"agencia": agencia, "numero_conta": numero_conta, "usuario": usuario}
  print("\nUsuário não encontrado, fluxo de criação de conta encerrado!")

def listar_contas(contas: str):
  for conta in contas:
    linha = f"""\
      Agência:\t{conta['agencia']}
      C/C:\t\t{conta['numero_conta']}
      Titular:\t{conta['usuario']['nome']}
    """
    print("=" * 100)
    print(linha)


def novo_usuario(usuarios: str):
  cpf = input("Informe o CPF (somente número): ")
  usuario = filtrar_usuario(cpf, usuarios)

  if usuario:
    print("Já existe usuário com esse CPF!")
    return

  nome = input("Informe o nome completo: ")
  data_nascimento = input("Informe a data de nascimento (dd-mm-aaaa): ")
  endereco = input("Informe o endereço (logradouro, nro - bairro - cidade/sigla estado): ")

  usuarios.append({"nome": nome, "data_nascimento": data_nascimento, "cpf": cpf, "endereco": endereco})

  print("Usuário criado com sucesso!")

def filtrar_usuario(cpf: str, usuarios: str):
  usuarios_filtrados = [usuario for usuario in usuarios if usuario["cpf"] == cpf]
  return usuarios_filtrados[0] if usuarios_filtrados else None


def main():
  agencia = "0001"
  saldo = 0
  saques_limite = 3
  valor_limite_saque = 500
  saques_realizados = 0
  extrato = []
  usuarios = []
  contas = []


  while True:
    opcao = input(menu())
    if opcao == "d":
      valor = float(input("Informe o valor do depósito: "))
      extrato, saldo = deposito(valor, saldo, extrato)

    elif opcao == "s":
      valor = float(input("Informe o valor do saque: "))
      extrato, saldo = saque(valor, saldo, saques_realizados, valor_limite_saque, saques_limite, extrato)

    elif opcao == "e":
      extrato_func(saldo, extrato)

    elif opcao == "nc":
      numero_conta = len(contas) + 1
      conta = nova_conta(agencia, numero_conta, usuarios)
      if conta:
        contas.append(conta)
    elif opcao == "lc":
      listar_contas(contas)

    elif opcao == "nu":
      novo_usuario(usuarios)

    elif opcao == "q":
      break
    else:
      print("Operação inválida, por favor selecione novamente a operação desejada.")


if __name__ == "__main__":
  main()
```
